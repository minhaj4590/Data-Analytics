# **Population Of Pakistan**

### **Description**
Pakistan Population Dataset provides comprehensive demographic information about the population of Pakistan. This dataset includes a wide range of population statistics, such as:

`Age Distribution:` Details about the population spread across various age groups, allowing for analysis of the age structure in different regions.

`Gender Distribution:` Insights into the gender composition within the population.

`Geographical Breakdown:` Data categorized by provinces and districts, offering a granular view of the population distribution across urban and rural areas.

`Urban and Rural Population:` Information on the division of population between urban and rural areas, which is critical for urban planning and resource allocation.

`Population Growth Trends:` Historical data that can be used to analyze trends in population growth over time.

`Years of data`: 1998, 2017

### **Data Source:**
This dataset is collected for analysis from this website: [kaggle.com](https://www.kaggle.com/datasets/iqramahmood004/pakistan-population-dataset)
