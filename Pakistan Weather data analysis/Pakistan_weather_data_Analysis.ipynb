However, I have very closely observed the values of NO2 throughout the data and it seems like the higher values of NO2 are associated with cold months (winter). So lets find out the answer to the third question 'Is there a relation between air polution and seasons?'

```python
# I will create a method to remove year from the Dates column where exists
# so we can focus on the dates alone.

# Only the 2019 data has this problem
def remove_year(x):
    return x.replace('-2019', '')
    
def remove_spaces(x):
    return x.replace(' ', '')

# Lets use the balanced data frame for this one not the filtered one
balanced_data['Date'] = balanced_data['Date'].apply(remove_year)
balanced_data['Date'] = balanced_data['Date'].apply(remove_spaces)
```

```python
# Lets get the season using month name

balanced_data['Season'] = np.nan 

def ident_season(df_str):
    month = df_str.split('-')[1]
    if month in ['Dec', 'Jan', 'Feb']:
        return 'Winter'
    elif month in ['Mar', 'Apr', 'May']:
        return 'Spring'
    elif month in ['Jun', 'Jul', 'Aug']:
        return 'Summer'
    elif month in ['Sep', 'Oct', 'Nov']:
        return 'Fall'
    else:
        return np.nan

balanced_data['Season'] = balanced_data['Date'].apply(ident_season)
```

```python
balanced_data.tail()
```

```python
balanced_data.loc[(balanced_data['Year'] > 2021) & (balanced_data['Season'] == 'Winter')].describe()
```

```python
data_by_season = balanced_data.groupby('Season').agg({
    'Temperature' : 'mean',
    'Humidity' : 'mean',
    'SO2' : 'mean', 
    'NO2' : 'mean', 
    'PM2.5' : 'mean'
})

data_by_season
```

Winters have significantly higher levels of air pollution compared to other seasons, this is likely due to
